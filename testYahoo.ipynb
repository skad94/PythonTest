```python
import yfinance as yf
```

```python
apple= yf.Ticker("aapl")
Facebook= yf.Ticker("META")
```

```python
# show actions (dividends, splits)
print(type(apple))
```

```python
# get all stock info
apple.info

# get historical market data
hist = apple.history(period="1mo")

# show meta information about the history (requires history() to be called first)
apple.history_metadata

# show actions (dividends, splits, capital gains)
apple.actions
apple.dividends
apple.splits
apple.capital_gains  # only for mutual funds & etfs

# show share count
apple.get_shares_full(start="2022-01-01", end=None)

# show financials:
# - income statement
apple.income_stmt
apple.quarterly_income_stmt
# - balance sheet
apple.balance_sheet
apple.quarterly_balance_sheet
# - cash flow statement
apple.cashflow
apple.quarterly_cashflow
# see `Ticker.get_income_stmt()` for more options

# show holders
apple.major_holders
apple.institutional_holders
apple.mutualfund_holders
apple.insider_transactions
apple.insider_purchases
apple.insider_roster_holders

# show recommendations
apple.recommendations
apple.recommendations_summary
apple.upgrades_downgrades

# Show future and historic earnings dates, returns at most next 4 quarters and last 8 quarters by default. 
# Note: If more are needed use apple.get_earnings_dates(limit=XX) with increased limit argument.
apple.earnings_dates

# show ISIN code - *experimental*
# ISIN = International Securities Identification Number
apple.isin

# show options expirations
apple.options

# show news
apple.news

# get option chain for specific expiration
#opt = apple.option_chain('2024-02-07')
# data available via: opt.calls, opt.puts
```
